```python
import matplotlib.pyplot as plt
import math

x = range(0, 10)
y = [n**2 - n + 2 for n in x]
plt.plot(x, y)
plt.savefig('my-figure')
plt.annotate('Origin', xy=(0, 0), xytext=(-3, 5),
             arrowprops={'facecolor': 'blue'})
plt.show()
```

```python
x = range(0, 5)

plt.figure(figsize=(14, 6)) # (width, height)

y = [math.sqrt(n) for n in x]
plt.plot(x, y, c='blue', label='$y=\sqrt{x}$')

y = [n**3 for n in x]
plt.plot(x, y, c='orange', label='O(log n)')

y = [2**n for n in x]
plt.plot(x, y, c='green', label='$y=2^x$')

y = [1/(n+1) for n in x]
plt.plot(x, y, c='red', label='$y= \dfrac{1}{x+1}$')


plt.legend(loc='upper right')
plt.title('Problems from question 4')
plt.xlabel('$x$')

plt.show()
```
