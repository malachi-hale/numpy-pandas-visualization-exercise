```python
import matplotlib.pyplot as plt
import math

x = range(0, 10)
y = [n**2 - n + 2 for n in x]
plt.plot(x, y)
plt.savefig('my-figure')
plt.annotate('Origin', xy=(0, 0), xytext=(-3, 5),
             arrowprops={'facecolor': 'blue'})
plt.show()
```

```python
x = range(0, 10)

ax1 = plt.subplot(221)
y = [math.sqrt(n) for n in x]
plt.plot(x, y)
ax1.title.set_text('$y=\sqrt{x}$')

ax2 = plt.subplot(222)
y = [n**3 for n in x]
plt.plot(x, y)
ax2.title.set_text('$y=x^3$')

ax3 =plt.subplot(223)
y = [2**n for n in x]
plt.plot(x, y)
ax3.title.set_text('$y=2^x$')

ax4 =plt.subplot(224)
y = [1/(n+1) for n in x]
plt.plot(x, y)
ax4.title.set_text('$y=\dfrac{1}{1+n}$')

plt.suptitle('Subplots')

plt.tight_layout()

plt.show()
```

```python
x = range(0, 4)

plt.figure(figsize=(14, 6)) 

y = [math.sqrt(n) for n in x]
plt.plot(x, y, c='blue', label='$y=\sqrt{x}$')

y = [n**3 for n in x]
plt.plot(x, y, c='orange', label='$y = n^3$')

y = [2**n for n in x]
plt.plot(x, y, c='green', label='$y=2^x$')

y = [1/(n+1) for n in x]
plt.plot(x, y, c='red', label='$y= \dfrac{1}{x+1}$')


plt.legend(loc='upper right')
plt.title('Problems from question 4')
plt.xlabel('$x$')

plt.show()
```
