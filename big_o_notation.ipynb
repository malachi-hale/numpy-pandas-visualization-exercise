```python
import matplotlib.pyplot as plt
import math

x = range(1, 4)

plt.figure(figsize=(14, 6)) 

y = [0*n for n in x]
plt.plot(x, y, c='blue', label='$O(1)$')

y = [math.log(n) for n in x]
plt.plot(x, y, c='orange', label='$O(\log n)$')

y = [n for n in x]
plt.plot(x, y, c='green', label='$O(n)$')

y = [n*math.log(n) for n in x]
plt.plot(x, y, c='red', label='$O(n*\log n$)')

y = [n**2 for n in x]
plt.plot(x, y, c='yellow', label='$O(n^2)$')

y = [math.factorial(n) for n in x]
plt.plot(x, y, c='purple', label='$O(n!)$')

y = [n**n for n in x]
plt.plot(x, y, c='pink', label='$O(n^n)$')




plt.legend(loc='upper right')
plt.title('Big O Notation')
plt.xlabel("Elements")
plt.ylabel("Operations")


plt.show()
```
